```python
df = pd.DataFrame({
    'Math':[85,78,932,88,76],
    'Science':[88,79,95,84,82],
    'English':[75,83,89,92,80]
})
correlation_matrix = df.corr()
print(correlation_matrix)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('cotrrelation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
math_counts = df['Math'].value_counts()
labels = [str(score) for score in math_counts.index]
math_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=labels,
    colors=['red', 'green', 'blue', 'yellow', 'orange'],
    title='Pie Chart: Math Score Distribution'
)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['Math'], label='Math', color='blue')
plt.title('Line Plot: Math Across Index')
plt.xlabel('Index')
plt.ylabel('Math')
plt.legend()
plt.grid(True)
plt.show()
```
